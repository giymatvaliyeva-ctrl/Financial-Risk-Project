```python
#4.	 Examine the distribution of employees across different departments and job roles.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
employee = pd.read_csv('Employee.csv')
dept_job_df = employee.groupby(['Department', 'JobRole'])['EmployeeID'].count().reset_index()
dept_job_df.columns = ['Department', 'JobRole', 'Employee_Count']
dept_job_df

#heatmap
heatmap_data = dept_job_df.pivot(index='JobRole', columns='Department', values='Employee_Count').fillna(0).astype(int)
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='Blues')
plt.title('Employee distribution across departments and job roles')
plt.ylabel('Job Role')
plt.xlabel('Department')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

#barplot
plt.figure(figsize=(12,6))
sns.barplot(data=dept_job_df, x='Department', y='Employee_Count', hue='JobRole', dodge=True)
plt.title('Employee distribution across departments and job roles')
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.xticks(rotation=30)
plt.legend(title='Job Role', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#6 Calculate average satisfaction levels for different groups (e.g., by department, gender, education level).
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

employee = pd.read_csv('Employee.csv')
performance = pd.read_csv("Performance Rating.csv")
performance.columns = performance.columns.str.strip()
employee.columns = employee.columns.str.strip()
# we try to merge EmployeeID tables
df = pd.merge(employee, performance, on='EmployeeID', how='left')
# we choose satisfaction columns
satisfaction_cols = ['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction', 'WorkLifeBalance']
# we calculate average satisfaction
df['Avg_Satisfaction'] = df[satisfaction_cols].mean(axis=1)
# we average satisfaction for department
dept_satisfaction = df.groupby('Department')['Avg_Satisfaction'].mean().reset_index()
dept_satisfaction.columns = ['Department', 'Avg_Satisfaction']
# we average satisfaction for gender
gender_satisfaction = df.groupby('Gender')['Avg_Satisfaction'].mean().reset_index()
gender_satisfaction.columns = ['Department', 'Avg_Satisfaction']
# we average satisfaction for education
education_satisfaction = df.groupby('Education')['Avg_Satisfaction'].mean().reset_index()
education_satisfaction.columns = ['Education_Level', 'Avg_Satisfaction']
#Department
plt.figure(figsize=(8,5))
sns.barplot(data=dept_satisfaction, x='Department', y='Avg_Satisfaction', color='skyblue')
plt.title('Average satisfaction for department')
plt.ylabel('Average Satisfaction')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

#Education
plt.figure(figsize=(8,5))
sns.barplot(data=education_satisfaction, x='Education_Level', y='Avg_Satisfaction', color='salmon')
plt.title('Average satisfaction for education')
plt.ylabel('Average Satisfaction')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()


```

```python
#7. Investigate the relationship between satisfaction levels and attrition.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

employee = pd.read_csv('Employee.csv')
performance = pd.read_csv("Performance Rating.csv")
#Clear column names
employee.columns = employee.columns.str.strip()
performance.columns = performance.columns.str.strip()
#Merge
df = pd.merge(employee, performance, on='EmployeeID', how='left')
#Satisfaction columns
satisfaction_cols = ['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction', 'WorkLifeBalance']
#Average satisfaction for every job
df['Avg_Satisfaction'] = df[satisfaction_cols].mean(axis=1)
#Average satisfaction with Attrition
attrition_satisfaction = df.groupby('Attrition')['Avg_Satisfaction'].mean().reset_index()
attrition_satisfaction.columns = ['Attrition', 'Avg_Satisfaction']
print(attrition_satisfaction)

#Barplot
plt.figure(figsize=(6,4))
sns.barplot(data=attrition_satisfaction, x='Attrition', y='Avg_Satisfaction', color='skyblue')
plt.title('Average satisfaction for attrition')
plt.ylabel('Average satisfaction')
plt.xlabel('Attrition')
plt.show()

#boxplot
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Attrition', y='Avg_Satisfaction', color='green')
plt.title('Average satisfaction for employment')
plt.ylabel('Average satisfaction')
plt.xlabel('Attrition')
plt.show()
```

```python
#10.Calculate average ratings by department and job role.

employee = pd.read_csv('Employee.csv')
performance = pd.read_csv("Performance Rating.csv")
# We clean up column names
employee.columns = employee.columns.str.strip()
performance.columns = performance.columns.str.strip()
# Merge
df = pd.merge(employee, performance, on='EmployeeID', how='left')
# Department and JobRole
dept_job_rating = df.groupby(['Department', 'JobRole'])[['SelfRating', 'ManagerRating']].mean().reset_index()
dept_job_rating

# Pivot for Heatmap
heatmap_self = dept_job_rating.pivot(index='JobRole', columns='Department', values='SelfRating').fillna(0)
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_self, annot=True, fmt=".2f", cmap='Blues')
plt.title('Average SelfRating by Department and JobRole')
plt.ylabel('JobRole')
plt.xlabel('Department')
plt.show()


# ManagerRating

heatmap_manager = dept_job_rating.pivot(index='JobRole', columns='Department', values='ManagerRating').fillna(0)
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_manager, annot=True, fmt=".2f", cmap='Greens')
plt.title('Average ManagerRating by Department and JobRole')
plt.ylabel('JobRole')
plt.xlabel('Department')
plt.show()
```

```python
#11.Create visualizations such as bar charts or histograms to show the distribution of satisfaction and rating levels.

employee = pd.read_csv('Employee.csv')
performance = pd.read_csv("Performance Rating.csv")

employee.columns = employee.columns.str.strip()
performance.columns = performance.columns.str.strip()
df = pd.merge(employee, performance, on='EmployeeID', how='left')
# Satisfaction and rating columns
satisfaction_cols = ['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction', 'WorkLifeBalance']
rating_cols = ['SelfRating', 'ManagerRating']
# Satisfaction columns
plt.figure(figsize=(12,8))
for i, col in enumerate(satisfaction_cols):
  plt.subplot(2,2,i+1)
  sns.histplot(df[col], bins=5, kde=False, color='skyblue')
  plt.title(f'{col} Distribution')
  plt.xlabel('Level')
  plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Rayting columns
plt.figure(figsize=(8,4))
for col in rating_cols:
  sns.histplot(df[col], bins=10, kde=True, label=col)
plt.title('Distribution of SelfRating and ManagerRating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.legend()
plt.show()

avg_satisfaction = df[satisfaction_cols].mean().reset_index()
avg_satisfaction.columns = ['Satisfaction_Type', 'Avg_Level']

plt.figure(figsize=(8,4))
sns.barplot(data=avg_satisfaction, x='Satisfaction_Type', y='Avg_Level', color='salmon')
plt.title('Average Satisfaction Levels')
plt.ylabel('Average Level')
plt.xticks(rotation=30)
plt.show()


```

```python
#12 Evaluate departmental performance, job satisfaction, and attrition rates.

employee = pd.read_csv('Employee.csv')
performance = pd.read_csv("Performance Rating.csv")

employee.column = employee.columns.str.strip()
performance.columns = performance.columns.str.strip()
df = pd.merge(employee, performance, on='EmployeeID', how='left')
# Satisfaction columns
satisfaction_cols = ['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction', 'WorkLifeBalance']
# Average satisfaction per employee
df['Avg_Satisfaction'] = df[satisfaction_cols].mean(axis=1)
# Average SelfRating, ManagerRating, satisfaction and attrition by department
dept_summary = df.groupby('Department').agg(
    Avg_SelfRating=('SelfRating','mean'),
    Avg_ManagerRating=('ManagerRating','mean'),
    Avg_Satisfaction=('Avg_Satisfaction','mean'),
    AttritionRate=('Attrition', lambda x: (x=='Yes').mean()*100) ).reset_index()
dept_summary
#Performance
plt.figure(figsize=(10,5))
sns.barplot(data=dept_summary, x='Department', y='Avg_SelfRating', color='skyblue', label='SelfRating')
sns.barplot(data=dept_summary, x='Department', y='Avg_ManagerRating', color='salmon', alpha=0.6, label='ManagerRating')
plt.title('Average Performance Rating by Department')
plt.ylabel('Average Rating')
plt.legend()
plt.xticks(rotation=30)
plt.show()

#Job Satisfaction
plt.figure(figsize=(10,5))
sns.barplot(data=dept_summary, x='Department', y='Avg_Satisfaction', color='lightgreen')
plt.title('Average Perfomance Satisfaction by Department')
plt.ylabel('Orta Satisfaction')
plt.xticks(rotation=30)
plt.show()
# Employee turnover rate
plt.figure(figsize=(10,5))
sns.barplot(data=dept_summary, x='Department', y='AttritionRate', color='orange')
plt.title('Department Attrition Rate (%)')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=30)
plt.show()

```

```python
#13 Identify high-performing and low-performing departments.
employee = pd.read_csv('Employee.csv')
performance = pd.read_csv("Performance Rating.csv")
employee.columns = employee.columns.str.strip()
performance.columns = performance.columns.str.strip()
df = pd.merge(employee, performance, on='EmployeeID', how='left')
# Average SelfRating and ManagerRating by department
dept_perf = df.groupby('Department')[['SelfRating', 'ManagerRating']].mean().reset_index()
dept_perf['Avg_Performance'] = dept_perf[['SelfRating','ManagerRating']].mean(axis=1)
# High performing department
high_perf_dept = dept_perf.loc[dept_perf['Avg_Performance'].idxmax()]
print("High performing department:")
print(high_perf_dept)
# Low performing department
low_perf_dept = dept_perf.loc[dept_perf['Avg_Performance'].idxmin()]
print("\nLow performing department:")
print(low_perf_dept)
#  List the departments
dept_perf_sorted = dept_perf.sort_values(by='Avg_Performance', ascending=False)
print("\nPerformance ranking of all departments:")
print(dept_perf_sorted)
```

```python
#14. Analyze gender distribution, salary differences, and diversity within the organization.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

employee = pd.read_csv('Employee.csv')
employee.columns = employee.columns.str.strip()
# Gender distribution
gender_counts = employee['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender','Count']
print("Gender distribution:")
gender_counts

plt.figure(figsize=(6,4))
sns.barplot(data=gender_counts, x='Gender', y='Count', color='skyblue')
plt.title('Gender Breakdown by Organization')
plt.ylabel('Number of Employees')
plt.show()

# Salary differences (by gender)
plt.figure(figsize=(6,4))
sns.boxplot(data=employee, x='Gender', y='Salary', color='lightgreen')
plt.title('Gender Wage Gap')
plt.ylabel('Monthly Income')
plt.xticks(rotation=30)
plt.show()
# Diversity: Gender vs Department
plt.figure(figsize=(10,6))
sns.countplot(data=employee, x='Department', hue='Gender')
plt.title('Gender Diversity by Department')
plt.ylabel('Number of employees')
plt.xticks(rotation=30)
plt.show()
#  Gender vs JobRole
plt.figure(figsize=(12,6))
sns.countplot(data=employee, x='JobRole', hue='Gender')
plt.title('Gender Diversity by JobRole')
plt.ylabel('Number of employees')
plt.xticks(rotation=45)
plt.show()


```

```python
#15. Evaluate whether gender or diversity plays a role in job satisfaction, performance, or attrition.

employee = pd.read_csv('Employee.csv')
performance = pd.read_csv('Performance Rating.csv')
employee.columns = employee.columns.str.strip()
performance.columns = performance.columns.str.strip()
df = pd.merge(employee, performance, on='EmployeeID', how='left')
satisfaction_cols = ['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction', 'WorkLifeBalance']
df['Avg_Satisfaction'] = df[satisfaction_cols].mean(axis=1)

import seaborn as sns
import matplotlib.pyplot as plt

# Gender vs Avg Satisfaction
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Gender', y='Avg_Satisfaction', color='lightblue')
plt.title('Average Job Satisfaction by Gender')
plt.ylabel('Average Satisfaction')
plt.show()

# Gender vs Performance
df_melted = df.melt(id_vars=['Gender'], value_vars=['SelfRating','ManagerRating'], var_name='RatingType', value_name='Rating')

plt.figure(figsize=(6,4))
sns.boxplot(data=df_melted, x='Gender', y='Rating', hue='RatingType', palette='Set1')
plt.title('Performance Rating by Gender')
plt.ylabel('Rating')
plt.show()

# Gender vs Attrition (employee turnover rate)
attrition_gender = df.groupby('Gender')['Attrition'].apply(lambda x: (x=='Yes').mean()*100).reset_index()
attrition_gender.columns = ['Gender','AttritionRate']

plt.figure(figsize=(6,4))
sns.barplot(data=attrition_gender, x='Gender', y='AttritionRate', color='lightcoral')  # palette əvəzinə color istifadə
plt.title('Employee Turnover Rate (%) by Gender')
plt.ylabel('Attrition Rate (%)')
plt.show()

#(Department + Gender) vs Avg Satisfaction
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Department', y='Avg_Satisfaction', hue='Gender', palette='Set2')
plt.title('Department and Gender by Job satisfaction')
plt.xticks(rotation=30)
plt.show()
#(Department + Gender) vs Performance
df_melted_dept = df.melt(id_vars=['Department','Gender'], value_vars=['SelfRating','ManagerRating'], var_name='RatingType', value_name='Rating')

plt.figure(figsize=(12,6))
sns.boxplot(data=df_melted_dept, x='Department', y='Rating', hue='Gender', palette='Set1')
plt.title('Department and Gender Performance')
plt.xticks(rotation=30)
plt.show()
```
